```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://www.espncricinfo.com/records/most-runs-in-career-284269'
```

```python
website = requests.get(url)
```

```python
soup = BeautifulSoup(website.text,'html')
```

```python
soup
```

#### view all tables

```python
table = soup.find_all('table', class_ = 'ds-w-full ds-table ds-table-xs ds-table-auto ds-w-full ds-overflow-scroll ds-scrollbar-hide')
table
```

```python
table_titles = []
for element in table:
    titles = element.find('tr')
    for title in titles:
        table_titles.append(title.text.strip())
        
```

```python
table_titles
```

```python
df = pd.DataFrame(columns = table_titles)
```

```python
df
```

```python
individual_data_list=[]
for element in table:
    individual_data = element.find_all('td')
    for title in individual_data:
        individual_data_list.append(title.text.strip())
```

```python
individual_data_list
```

```python
new_list = individual_data_list[15:]
```

```python
new_list
```

```python
a=[]
for x in new_list:
    a.append(x)
    
    if len(a) == 15:
        length = len(df)
        df.loc[length] = a
        a=[]
        
```
